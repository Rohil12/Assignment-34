```python
#Q.1
import scipy.stats as sp
import numpy as np

mean = 50
std = 5
n = 30

error = std / np.sqrt(n)

z_test = sp.norm.ppf(0.975)

margin= error * z_test

confidence_interval = (mean - margin , mean + margin)

print("95% confidence interval:",confidence_interval)
```

```python
# Q.2
import scipy.stats as sp
import numpy as np

observed = np.array([20,25,15,10,15,15])

expected_proportions = np.array([0.20,0.20,0.20,0.10,0.10,0.20])
total_count = np.sum(observed)
expected = expected_proportions * total_count

chi2,p_value = sp.chisquare(f_obs=observed,f_exp=expected)

print(f"Chi-squared statistics:{chi2}")
print(f"P-value:{p_value}")

significance_level = 0.05
if p_value < significance_level:
    print("Reject the null hypothesis: The observed distribution is significantly different from the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed distribution is not significantly different from the expected distribution.")

```

```python
# Q.4
import scipy.stats as stats

# Given data
population_size = 500
num_smokers = 60

# Calculate sample proportion
sample_proportion = num_smokers / population_size

# Calculate z-score for 95% confidence level
z_score = stats.norm.ppf(0.975)  # 97.5th percentile of the standard normal distribution

# Calculate confidence interval
margin_of_error = z_score * (sample_proportion * (1 - sample_proportion) / population_size)**0.5
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

# Print the confidence interval
print("95% Confidence Interval for the true proportion of smokers:")
print(f"{lower_bound:.3f} to {upper_bound:.3f}")
```

```python
# Q.6
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Parameters
df = 10  # degrees of freedom
x = np.linspace(0, 30, 1000)  # chi-square range

# Chi-square distribution
y = chi2.pdf(x, df)

# Plotting the chi-square distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='Chi-square Distribution')

# Shading the area for chi-square statistic of 15
x_fill = np.linspace(0, 15, 1000)
y_fill = chi2.pdf(x_fill, df)
plt.fill_between(x_fill, y_fill, alpha=0.5, color='gray')

# Labels and title
plt.title('Chi-Square Distribution with 10 Degrees of Freedom')
plt.xlabel('Chi-Square Statistic')
plt.ylabel('Probability Density')
plt.legend()

# Show plot
plt.show()

```

```python
# Q.7
import scipy.stats as stats

sample = 1000
coke = 520 
proportion = coke/sample

z_score = stats.norm.ppf(0.99)

margin = z_score * proportion * (1-proportion/sample)**0.5
lower= proportion - margin
upper= proportion + margin

print("95% Confidence Interval for the true proportion of smokers:")
print(f"{lower:.3f} to {upper:.3f}")
```

```python
# Q.8
from scipy.stats import chisquare

# Observed and expected frequencies
observed = [45, 55]
expected = [50, 50]

# Chi-square test
chi2_stat, p_val = chisquare(observed, f_exp=expected)

print("Chi-square Statistic:", chi2_stat)
print("P-value:", p_val)

# Interpretation
alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: The coin is biased towards tails.")
else:
    print("Fail to reject the null hypothesis: No evidence that the coin is biased towards tails.")
```

```python
#Q.9
import numpy as np
import scipy.stats as stats

# Observed frequencies
observed = np.array([[60, 140], 
                     [30, 170]])

# Perform the chi-square test for independence
chi2_stat, p_val, dof, expected = stats.chi2_contingency(observed)

print("Chi-square Statistic:", chi2_stat)
print("P-value:", p_val)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:", expected)

# Interpretation
alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("Fail to reject the null hypothesis: No significant association between smoking status and lung cancer diagnosis.")
```

```python
import scipy.stats as stats

# Given values
sample_mean = 72
pop_mean = 70
sample_std = 10
sample_size = 30
alpha = 0.05

# Calculate the t-test statistic
t_statistic = (sample_mean - pop_mean) / (sample_std / (sample_size ** 0.5))

# Degrees of freedom
df = sample_size - 1

# Calculate the p-value
p_value = 2 * stats.t.sf(abs(t_statistic), df)

print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Decision
if p_value < alpha:
    print("Reject the null hypothesis: The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the population mean and 70.")
```
